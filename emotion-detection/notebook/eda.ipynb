```python
import os
import glob
from PIL import Image

base_dir = "../images/train"

total_greyscale_pixels = 0
total_rgb_pixels = 0
greyscale_count = 0
rgb_count = 0

for child_dir in os.listdir(base_dir):
    curr_dir = os.path.join(base_dir, child_dir)
    
    if not os.path.isdir(curr_dir):
        continue
    
    files = glob.glob(os.path.join(curr_dir, '*.jpg'))
    
    for file in files:
        img = Image.open(file)
        w, h = img.size
        pixels = w * h
        
        if img.mode == 'L':
            greyscale_count += 1
            total_greyscale_pixels += pixels
        elif img.mode in ['RGB', 'RGBA']:
            rgb_count += 1
            total_rgb_pixels += pixels

# Calculate memory
# Grayscale: 1 channel
greyscale_mb_uint8 = (total_greyscale_pixels * 1) / (1024 ** 2)
greyscale_mb_float32 = (total_greyscale_pixels * 1 * 4) / (1024 ** 2)

# RGB: 3 channels
rgb_mb_uint8 = (total_rgb_pixels * 3) / (1024 ** 2)
rgb_mb_float32 = (total_rgb_pixels * 3 * 4) / (1024 ** 2)

# Total
total_mb_uint8 = greyscale_mb_uint8 + rgb_mb_uint8
total_mb_float32 = greyscale_mb_float32 + rgb_mb_float32

# If all converted to RGB float32 (for PyTorch)
all_rgb_float32 = ((total_greyscale_pixels + total_rgb_pixels) * 3 * 4) / (1024 ** 2)

print(f"Greyscale images: {greyscale_count} ({total_greyscale_pixels:,} pixels)")
print(f"RGB images: {rgb_count} ({total_rgb_pixels:,} pixels)")
print(f"\n{'='*50}")
print(f"Memory as-is:")
print(f"  uint8 (raw): {total_mb_uint8:.2f} MB")
print(f"  float32 (PyTorch): {total_mb_float32:.2f} MB")
print(f"\nMemory if all converted to RGB float32:")
print(f"  {all_rgb_float32:.2f} MB")
print(f"  Per image: {all_rgb_float32 / (greyscale_count + rgb_count):.3f} MB")
```
